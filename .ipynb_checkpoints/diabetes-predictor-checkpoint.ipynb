## Parameters

```python
path_to_data = "diabetes_prediction_dataset.csv"
target_attr = "diabetes"
```

## Clean Data

```python
# filter out "no info" smoking history
diabetes_df = diabetes_df[diabetes_df["smoking_history"] != 'No Info']
```

```python
# change ever to never
diabetes_df["smoking_history"] = diabetes_df["smoking_history"].apply(lambda x: "never" if x == "ever" else x)

# change ever to never
diabetes_df["smoking_history"] = diabetes_df["smoking_history"].apply(lambda x: "former" if x == "not current" else x)
```

## Random Forest Model

```python
# create predictors
diabetes_df["smoking_history_code"] = diabetes_df["smoking_history"].astype("category").cat.codes
```

```python
diabetes_df["gender_code"] = diabetes_df["gender"].astype("category").cat.codes
```

```python
predictors = ["gender_code", "age", "smoking_history_code", "hypertension", "bmi"]
```

```python
# split dataset into 70 % training dataset and 30 % test datasets
testing = diabetes_df.sample(frac = 0.3, random_state = 42)
training = diabetes_df.drop(testing.index)
```

```python
# create a RF classifier
rf = RandomForestClassifier(n_estimators = 100, min_samples_split=10, random_state=1)

rf.fit(training[predictors], training[target_attr])

# performing predictions on the test dataset
preds = rf.predict(testing[predictors])
```

## Accuracy of Model

```python
acc = accuracy_score(testing[target_attr], preds)
```

```python
acc
```

```python
# examine accuracy
combined = pd.DataFrame(dict(actual=testing[target_attr], prediction=preds))
```

```python
# confusion matrix
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

```python
precision_score(testing[target_attr], preds)
```

## Improve Machine Learning Model

```python
additional_predictors = ["gender_code", "age", "smoking_history_code", "hypertension", "bmi", "heart_disease"]
```

```python
# split dataset into 70 % training dataset and 30 % test datasets
testing_B = diabetes_df.sample(frac = 0.3, random_state = 42)
training_B = diabetes_df.drop(testing_B.index)
```

```python
# create a RF classifier
rf_B = RandomForestClassifier(n_estimators = 100, min_samples_split=10, random_state=1)

rf_B.fit(training_B[additional_predictors], training_B[target_attr])

# performing predictions on the test dataset
preds_B = rf_B.predict(testing_B[additional_predictors])
```

```python
acc_B = accuracy_score(testing_B[target_attr], preds_B)
acc
```

```python
precision_score(testing_B[target_attr], preds_B)
```

## Hyperparameter Tuning

```python
param_dist = {'n_estimators': randint(50,500),
              'max_depth': randint(1,20)}

# Create a random forest classifier
rf_C = RandomForestClassifier()

# Use random search to find the best hyperparameters
rand_search = RandomizedSearchCV(rf_C, 
                                 param_distributions = param_dist, 
                                 n_iter=5, 
                                 cv=5)

# Fit the random search object to the data
rand_search.fit(training_B[additional_predictors], training_B[target_attr])
```

```python
best_rf = rand_search.best_estimator_

# Print the best hyperparameters
print('Best hyperparameters:',  rand_search.best_params_)
```

```python
preds_C = best_rf.predict(testing_B[additional_predictors])
```

```python
acc_C = accuracy_score(testing_B[target_attr], preds_C)
acc
```

```python
precision_score(testing_B[target_attr], preds_C)
```
